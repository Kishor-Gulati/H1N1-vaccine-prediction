# Vaccine Usage Prediction

### Abstract:
Subjects receiving the same vaccine often show different levels of immune responses and some may even present adverse side effects to the vaccine. Systems vaccinology can combine data and machine learning techniques to obtain highly predictive signatures of vaccine immunogenicity and reactogenicity. Currently, several machine learning methods are already available to researchers with no background in bioinformatics.

### Problem Statement:
Predict how likely it is that the people will take an H1N1 flu vaccine using Logistic Regression.

### Dataset Information:
* <u>unique_id</u>:- Unique identifier for each respondent.
* <u>h1n1_worry</u>:- Worry about the h1n1 flu - (0,1,2,3) - (0=Not worried at all, 1=Not very worried, 2=Somewhat worried, 3=Very worried).
* <u>h1n1_awareness</u>:- Signifies the amount of knowledge or understanding the respondent has about h1n1 flu - (0,1,2) - (0=No knowledge, 1=little knowledge, 2=good knowledge.)
* <u>antiviral_medication</u>:- Has the respondent taken antiviral vaccination - (0,1).
* <u>contact_avoidance</u>:- Has avoided any close contact with people who have flu-like symptoms - (0,1).
* <u>bought_face_mask</u>:- Has the respondent bought mask or not - (0,1).
* <u>wash_hands_frequently</u>:- Washes hands frequently or uses hand sanitizer - (0,1).
* <u>avoid_large_gatherings</u>:- Has the respondent reduced time spent at large gatherings - (0,1).
* <u>reduced_outside_home_cont</u>:- Has the respondent reduced contact with people outside their own house - (0,1).
* <u>avoid_touch_face</u>:- Avoids touching nose, eyes, mouth - (0,1).
* <u>dr_recc_h1n1_vacc</u>:- Doctor has recommended h1n1 vaccine - (0,1).
* <u>dr_recc_seasonal_vacc</u>:- The doctor has recommended seasonal flu vaccine - (0,1).
* <u>chronic_medic_condition</u>:- Has any chronic medical condition - (0,1).
* <u>cont_child_undr_6_mnth</u>:- Has regular contact with child the age of 6 months - (0,1).
* <u>is_health_worker</u>:- Is respondent a health worker - (0,1).
* <u>has_health_insur</u>:- Does respondent have health insurance - (0,1).
* <u>is_h1n1_vacc_effective</u>:- Does respondent think that the h1n1 vaccine is effective - (1,2,3,4,5) - (1=Thinks not effective at all, 2=Thinks it is not very effective, 3=Doesn't know if it is effective or not, 4=Thinks it is somewhat effective, 5=Thinks it is highly effective).
* <u>is_h1n1_risky</u>:- What respondents think about the risk of getting ill with h1n1 in the absence of the vaccine- (1,2,3,4,5) - (1=Thinks it is not very low risk, 2=Thinks it is somewhat low risk, 3=don’t know if it is risky or not, 4=Thinks it is a somewhat high risk, 5=Thinks it is very highly risky).
* <u>sick_from_h1n1_vacc</u>:- Does respondent worry about getting sick by taking the h1n1 vaccine - (1,2,3,4,5) - (1=Respondent not worried at all, 2=Respondent is not very worried, 3=Doesn't know, 4=Respondent is somewhat worried, 5Respondent is very worried).
* <u>is_seas_vacc_effective</u>:- Does respondent think that the seasonal vaccine is effective- (1,2,3,4,5) - (1=Thinks not effective at all, 2=Thinks it is not very effective, 3=Doesn't know if it is effective or not, 4=Thinks it is somewhat effective, 5=Thinks it is highly effective).
* <u>is_seas_flu_risky</u>:- What respondents think about the risk of getting ill with seasonal flu in the absence of the vaccine- (1,2,3,4,5) - (1=Thinks it is not very low risk, 2=Thinks it is somewhat low risk, 3=Doesn't know if it is risky or not, 4=Thinks it is somewhat high risk, 5=Thinks it is very highly risky).
* <u>sick_from_seas_vacc</u>:- Does respondent worry about getting sick by taking the seasonal flu vaccine - (1,2,3,4,5) - (1=Respondent not worried at all, 2=Respondent is not very worried, 3=Doesn't know, 4=Respondent is somewhat worried, 5=Respondent is very worried).
* <u>age_bracket</u>:- Age bracket of the respondent - (18 - 34 Years, 35 - 44 Years, 45 - 54 Years, 55 - 64 Years, 64+ Years).
* <u>qualification</u>:- Qualification/education level of the respondent as per their response -(<12 Years, 12 Years, College Graduate, Some College).
* <u>race</u>:- Respondent's race - (White, Black, Other or Multiple, Hispanic).
* <u>sex</u>:- Respondent's sex - (Female, Male).
* <u>income_level</u>:- Annual income of the respondent as per the 2008 poverty Census - (<=75000−AbovePoverty, >75000, Below Poverty).
* <u>marital_status</u>:- Respondent's marital status - (Not Married, Married).
* <u>housing_status</u>:- Respondent's housing status - (Own, Rent).
* <u>employment</u>:- Respondent's employment status - (Not in Labor Force, Employed, Unemployed).
* <u>census_msa</u>:- Residence of the respondent with the MSA(metropolitan statistical area)(Non-MSA, MSA- Not Principle, CityMSA-Principle city).
* <u>no_of_adults</u>:- Number of adults in the respondent's house (0,1,2,3).
* <u>no_of_children</u>:- Number of children in the respondent's house (0,1,2,3).
* <u>h1n1_vaccine</u>:- (Dependent variable) Did the respondent receive the h1n1 vaccine or not(1,0) - (Yes, No).

### Pre-processing the data

```python
# Create a copy of the dataframe.
df_rev = pd.DataFrame.copy(df)
```

```python
df_rev.isnull().sum()
```

```python
# We see that "has_health_insur" field has a very high number of null values. We will find the ratio.
(df_rev["has_health_insur"].isnull().sum()/len(df_rev))*100
```

```python
# We do have null values in our dataset. We will now find the ratio of null values.

print("The ratio of 'records having any null values' to 'total records' = ",
      round(((len(df_rev)-len(df_rev.dropna()))/len(df_rev)*100),2), "%.")

print("The ratio of 'records having 30 null values' to 'total records' = ",
      round(((len(df_rev)-len(df_rev.dropna(thresh=30)))/len(df_rev)*100),2), "%.")
```

* We have a large number of records having any number of null values.
* Dropping them will cause a 55.84% data loss, which is not acceptable.
* We find that we have 6.57% records with a threshold of 30 and we can affoard to drop them.
* We also see that "has_health_insur" field has almost 46% data missing and hence, we drop the column.
* We will fill the remaining null data with the use of central tendencies.

```python
df_rev.dropna(thresh=30, inplace=True)
df_rev.drop("has_health_insur", axis=1, inplace=True)

# We confirm using .shape
print(df_rev.shape)
```

```python
# Replacing the missing values with their respective mode values
for value in df_rev.columns:
    df_rev[value].fillna(df_rev[value].mode()[0], inplace = True)

# We confirm that there are no more missing/NAN values in our dataframe.
df_rev.isnull().sum()
```

## Converting the categorical data into numerical data/values.
* We need to convert the categorical features to hold numeric values.
* We also need to convert the discrete numeric float values to integer.

```python
colname = []
for x in df_rev.columns:
    if df_rev[x].dtype == 'object':
        colname.append(x)
colname
```

```python
#from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for x in colname:
    df_rev[x] = le.fit_transform(df_rev[x])
    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature', x, ":")
    print('mapping', le_name_mapping)
```

### Splitting data into X(Independent) and Y(Dependent) variables

```python
X = df_rev.values[:, 0:-1]
Y = df_rev.values[:,-1]

# We confirm the shape of X and Y
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

We do not need to scale/tranform our data as we have all discrete numeric values with a low range.

### Running a basic model

```python
# Split the data into test and train

#from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_test:", Y_test.shape)
```

```python
#from sklearn.linear_model import LogisticRegression

# Create a model
classifier = LogisticRegression()

# fitting training data to the model
classifier.fit(X_train, Y_train)

# Predicting using .predict
Y_pred = classifier.predict(X_test)
print(Y_pred)
```

#### Checking the intercept and slope coefficients of our features

```python
print("Slope coef: ", classifier.intercept_)

temp = list(zip(df_rev.columns[:-1], classifier.coef_.ravel()))
temp_df = pd.DataFrame(temp, columns = ['Features', 'Coefficients'])
temp_df.style.hide_index()
```

#### Finding accuracy and metrics

```python
#from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print("Confusion Matrix: \n\n", confusion_matrix(Y_test, Y_pred)) 
print("\n Accuracy of the model: ", accuracy_score(Y_test, Y_pred))
print("\n\n Classification Report: \n\n", classification_report(Y_test, Y_pred))
```

```python
# Predicting using .predict_proba and storing the predicted probabilities

y_pred_prob = classifier.predict_proba(X_test)
print(y_pred_prob)
```

### Adjusting the threshold

```python
# printing type I and type II errors for every threshold in range (0.4-0.6)

least_error = []
type_1 = []
type_2 = []

for a in np.arange(0.4,0.61,0.01):
    predict_mine = np.where(y_pred_prob[:,1] > a, 1, 0)
    cfm = confusion_matrix(Y_test, predict_mine)
    total_err = cfm[0,1]+cfm[1,0]
    type_1.append(cfm[0,1])
    type_2.append(cfm[1,0])
    least_error.append(total_err)
    print("Errors at threshold ", a, ":", total_err, " , type 2 error:", cfm[1,0]," , type 1 error :", cfm[0,1])
```

```python
# Sorting our errors in ascending order
least_error.sort()
type_2.sort()
type_1.sort()

# Displaying errors in ascending order

print("Total errors:", least_error, "\n")
print("Type 2 errors:", type_2, "\n")
print("Type 1 errors:", type_1, "\n")

# Finding least errors

print("Least total error:", min(least_error))
print("Least type 2 error:", min(type_2))
print("Least type 1 error:", min(type_1))
```

### Using cross validation - K-Fold

```python
classifier=LogisticRegression()

# Performing kfold_cross_validation

#from sklearn.model_selection import KFold
kfold_cv=KFold(n_splits=10,random_state=10,shuffle=True)
#print(kfold_cv)

# Running the model using scoring metric as accuracy

#from sklearn.model_selection import cross_val_score
kfold_cv_result=cross_val_score(estimator=classifier,X=X_train, y=Y_train, cv=kfold_cv)

print("K-fold cross-validation Scores: \n",kfold_cv_result)  # We get 10 accuracy scores

# Finding the mean
print("\n K-fold cross-validation mean:", kfold_cv_result.mean())  # Avg of all 10 accuracy scores
```

### Model tuning

```python
for train_value, test_value in kfold_cv.split(X_train,Y_train):
    classifier.fit(X_train[train_value], Y_train[train_value]).predict(X_train[test_value])

    
Y_pred=classifier.predict(X_test)
# print(list(zip(Y_test,Y_pred)))

cfm = confusion_matrix(Y_test, Y_pred)  # Always pass the Y_test(Y_actuals) and then Y_pred.
print("Confusion Matrix: \n\n", cfm)

acc = accuracy_score(Y_test,Y_pred)
print("\n Accuracy of the model: ", acc)

Report = classification_report(Y_test,Y_pred)
print("\n\n Classification Report: \n\n", Report)
```

### Using cross validation - Stratified K-Fold

```python
classifier=LogisticRegression()

# Performing stratified kfold_cross_validation

#from sklearn.model_selection import StratifiedKFold
kfold_cv=StratifiedKFold(n_splits=10,random_state=10,shuffle=True)   # n_splits = 10 is k=10
# print(kfold_cv)

# Running the model using scoring metric as accuracy

#from sklearn.model_selection import cross_val_score
kfold_cv_result=cross_val_score(estimator=classifier,X=X_train, y=Y_train, cv=kfold_cv) # X capital , y lower case

print("Stratified K-fold cross-validation Scores: \n",kfold_cv_result)  # We get 10 accuracy scores

# Finding the mean
print("\n Stratified K-fold cross-validation mean:", kfold_cv_result.mean())  # Avg of all 10 accuracy scores
```

```python
for train_value, test_value in kfold_cv.split(X=X_train,y=Y_train):
    classifier.fit(X_train[train_value], Y_train[train_value]).predict(X_train[test_value])

    
Y_pred=classifier.predict(X_test)
# print(list(zip(Y_test,Y_pred)))

cfm = confusion_matrix(Y_test, Y_pred)  # Always pass the Y_test(Y_actuals) and then Y_pred.
print("Confusion Matrix: \n\n", cfm)

acc = accuracy_score(Y_test,Y_pred)
print("\n Accuracy of the model: ", acc)

Report = classification_report(Y_test,Y_pred)
print("\n\n Classification Report: \n\n", Report)
```
